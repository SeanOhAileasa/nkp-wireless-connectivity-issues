###### 1. Reflection, Refraction, and Latency
***

After completing this topic, you should be able to: ``given a scenario toubleshoot reflection, refraction and latency``

Now in this section we'll take a look at some of the issues that can arise with connectivity in wireless networks. 

And consider that, of course, wireless signals are sent using radio frequencies. And radio, in and of itself, is just another portion of the spectrum of light. 

So the problems that can arise are the same issues that we can see with visible light, just to a different degree. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Reflection
***

So beginning with reflection, wireless signals can, of course, bounce off certain surface types. And depending on the surface the degree of bounce will vary. 

For example, of course, if we shine a flash light on a mirror, almost all of the light reflects off. Whereas if you shine it on a wall, not nearly as much will reflect. 

So signals can be weakened, and this can cause interference depending on how much reflection there is. 

So in a practical scenario, you probably would not want a lot of glass, for example, around your wireless access points. 

The less, the better, because you want that signal to be as unimpeded as possible from the transmission point to the reception point. If there's absolutely nothing in between, that is your best case scenario. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Refraction
***

Now refraction is the bending of a signal wave as it changes speeds on a new medium. 

Now, of course, as it's traveling through the air and if it is unimpeded, then it's not going to change the medium. 

But, of course, in many cases, the signal has to travel through walls, and they can be made of different materials such as glass as just mentioned, concrete, wood and dry wall, metal. 

So as those physical types of media change, so can the refractive index, and this can cause lower data rates and reduced capacity.

So, again, ideally what you want is nothing in the way, but that's just not always an option. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Diffraction
***

And diffraction occurs when waves have to travel, if you will, around obstacles. 

Now, in many cases depending on what the obstacle is, it might travel through it. 

But something like a large hill, for example, is going to effectively split the wave. And it will kind of travel out on either side of the hill, so that if you were directly behind the hill, you wouldn't get very much of the signal at all. 

So the diffraction, of course, can change the wave direction and or the intensity. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Scattering
***

Now scattering is similar to refraction, but it's a little more unpredictable. 

And it's due to things such as dust or high humidity, or, again, inconsistencies of any given material through which it is traveling. 

And this can alter the overall signal integrity and the strength. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Absorption
***

And finally absorption, this is a common reaction when wireless signals meet different materials, again such as wood or concrete. 

Now, this isn't referring to how it scatters or diffuses, but rather how that material absorbs that energy. 

It is after all just a form of energy, so what happens is it gets transferred into heat. 

And different materials will transfer that energy differently, so it affects the overall signal strength. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Conclusion - Signal Strength and Quality
***

So ultimately, your best bet, as mentioned, is to have nothing in your way, but that's just not always an option. So whatever you can do to reduce the amount of obstacles and or try to keep it to materials that are fairly consistent, then the better the overall signal strength and quality will be.

###### 3. Incorrect Antenna Types and Placement
***

After completing this topic, you should be able to: ``troubleshoot incorrect antenna types and antenna placements``

In this presentation we'll talk about some issues that you may encounter with your antennas in a wireless environment. 

Beginning with the type, you might encounter simply the wrong type depending on what it is you're trying to achieve. 

There are different Wi-Fi antennas and each is designed for a specific purpose. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Directional versus Omnidirectional Antenna
***

Most notably, the directional antenna, which is designed to focus on certain areas of coverage, versus the omnidirectional antenna, which is designed for 360-degree coverage. 

So in some cases, you might know that you only want wireless in this particular area of a building for example. And in that case, you would want the directional antenna. 

But if it's something like a hotel, or an airport, or a public library or something along those lines where it's fairly large and fairly open. And you want everyone to be able to have access regardless of where they are, then of course you need the omnidirectional. 

So again, just ensure that you have the correct type for whatever it is you are trying to achieve. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Incorrect Antenna Aiming
***

Now there can also be issues with aiming, and this is a little more common on some of the legacy access points that used what are known as diversity antennas. And this is essentially just trading off one antenna to the other. When it detects a packet, it will try to negotiate the best signal over one antenna versus the other. 

And the position, and/or the aiming of those antennas can actually make a difference. 

Now, I do suggest you do a little bit of research on the position. I have seen conflicting statements that say you should position them like this, or like that. And it tends to change from time to time, and again, there are several other factors that can, of course, interfere with any given signal. So you always have to take that into consideration as well. 

But if they are improperly aligned, it could be detrimental to signal performance.

So perhaps a little bit of trial and error might be the best solution. 

But I've seen many documents say that both antennas should be pointing straight up. I've seen as many that say one should be pointing straight up and one should be pointing horizontally. Almost perpendicular to it, so the two antenna make an L if you will, and a few other configurations. So again, perhaps a little trial and error might be the best option there. 

But another possible issue is what's known as hidden node. And this occurs when two clients in opposite directions connect to the same access point but effectively cannot detect each other. 

So in other words, if you just imagine a circle then the access point is in the middle. Then one client is completely on the left side at the maximum range and the other client is completely on the right side at the maximum range. Each of them within range of the access point but not really within range of each other. Now, that doesn't really matter per say in terms of being able to exchange a packet, because they both still go through the access point. But recall that this is still an Ethernet network, and Ethernet specifies that two systems cannot send a packet over the same medium at the same time. Now, the medium in this case is just the air but you are still sending packets over the air at the same time in many cases and they quite simply will collide. You will still get collisions in wireless networks, so the fact that client A doesn't really see client B in terms of their distance. In essence means that they will both try to communicate at the same time with the access point, and in many cases this will actually cause a collision. Because of course the transmissions, event though they might be weak, they will still meet up with each other and that's a collision. So that, of course, will degrade the performance of the network. 

So ideally, you want all of the systems to be able to see the access point and to see each other, because then they can do what Ethernet does. 

They can listen, they can implement carrier sense multiple access with collision detection and collision avoidance. And they can see that system B is transmitting right now, so I will hold off and system B does the same thing. It sees that system A is transmitting so it will hold off so that these collisions of course, do not occur.

But if you do have something like a spectrum analyzer and you can get a very good picture of what the wireless signal is going to look like, then you can hopefully avoid some of these problems before they begin.

###### Service Set Identifier (SSID)
***

Now the SSID is quite simply a network-naming scheme. 

Something that you can call the wireless network. 

So this is a configurable parameter and it needs to be shared by the client and the access point. 

And more importantly, it needs to be consistent on both. 

So a misconfigured SSID, in essence, will prevent the client from being able to connect. Or they might even connect to the wrong network entirely if, for example, they just are typing in the name and they just made a simple typo. But yet there was another wireless within range that matched that name, they might actually connect to it instead of the intended one. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Passphrase
***

Now with respect to the passphrase, this is like any other password, you have to supply the correct one to be able to connect.

Now it does depend on the security that you have implemented with respect to what the passphrase will look like.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Wired Equivalent Privacy (WEP) Protocol
***

So if, for example, WEP is being used, the wired equivalent privacy protocol. Bear in mind this is not considered to be a secure Wi-Fi protocol by today's standards but it is backward compatible. 

So it's not gone, but the passphrase will vary depending on the strength of the encryption. 

At the lower 40 to 64-bit setting, you'll see a 10 hexadecimal character or 5 ASCII character passphrase. 

And if it's the higher setting of 104 up to 128, then you see 26 hexadecimal characters or 13 ASCII characters. 

So that in and of itself actually decreases the security a little bit. 

Because an attacker knows exactly how many characters to deal with, with respect to the passphrase. 

So again, WEP is not considered to be secure. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### WPA/WPA2 Personal
***

But if you see WPA or WPA2, there is the personal edition as well as the enterprise that we'll see in a moment, but this uses much stronger encryption than WEP. 

It uses a combination of the temporal key integrity protocol and the advanced encryption services to provide much stronger encryption. 

And it supports a preshared key of between 8 and 63 characters, and it can be any value in between. 

And so again, attackers do not know how many characters they're dealing with, whereas they do with WEP. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### WPA/WPA2 Enterprise
***

And in the enterprise edition, this is often implemented in business environments and provides even more security. 

Although it can be a little more complicated to set up. 

But users are assigned login credentials by administration. 

And those credentials must be presented at logon, and they can be revoked at any time by administration. And in essence, those connectivity credentials are provided to you by administration, and an attacker just, of course, has no idea what the credentials are supposed to look like.

So it's even more difficult, but the idea behind Enterprise is to allow an administrator to centrally configure all of the connectivity properties and then effectively distribute those or push them out to the clients. So that everyone has their own specific settings with respect to those credentials.

But ultimately, if the passphrase is incorrect, then you clearly will not be able to connect. So if all of a sudden it seems like you can't connect anymore, then it's very likely the passphrase has been changed on the access point, but your system has not been updated. So of course you need to make that change as soon as possible.

###### 8. Power Levels and Signal-to-noise Ratio

***
###### Power Level
***

Another issue that might arise in a wireless implementation is actually the power level of the access point itself. 

Now this, for the most part, is something that you only encounter when attempting to configure wireless access for very large environments, such as warehouses or airports. 

They do have the ability to adjust the power level. And this can negatively affect the coverage and the performance. 

In other words, actually having full power on a device can be problematic. Now, again, it's certainly not guaranteed but it's something that you want to be mindful of. 

Because what happens is that the power level is raised so that clients at a greater distance are able to connect. 

But the problem can in fact be that clients that are actually closer will pick up on that stronger signal and try to associate with that one even though it's farther away. 

So effectively they're trying to reach too far, if you will, rather than connecting to the one that they see right there. They see this signal that comes in and appears to be stronger and they think well, maybe I'll connect to that one instead. 

But then if they move even just a few feet further away, now they're really taxing the limit of that particular signal and they lose the signal. 

So, in essence, they should really remain connected to the one that is closest. 

So this can require some trial and error. And, in essence, determining what the best power level ultimately might be. 

But, surprisingly, you generally don't want to go straight for 100%, if you will. Maybe pull it back to 75 or 60 and just try some configurations and ensure that you are moving around. 

So that you can determine whether or not clients are actually seeing what appears to be a stronger signal. But, of course, is then very vulnerable to just moving a little bit farther away.

***
###### Signal-to-Noise Ratio (SNR)
***

So another consideration that the you might want to test for is what's known as the signal to noise ratio, or SNR. 

This is a measure of the quality of the signal. 

And a low SNR can indicate that there is an issue with background noise. In other words, you're getting more noise than signal. So some of the sources of noise can include microwave ovens, wireless devices, fluorescent lights even. But basically anything that can interfere with a radio signal. 

So ideally you want as little of that as possible. You want, in a perfect world, 100% signal and 0% noise. But, you know, that's probably not ever going to happen. 

So the higher the ratio, the better. And ultimately that means you are getting more signal than noise. And clearly that is what you want for your best quality connections.
